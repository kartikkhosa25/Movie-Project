```python
import pandas as pd
```

```python
movie = pd.DataFrame(pd.read_csv("/content/IMDb movies.csv"))
```

```python
movie.head()
```

```python
movie.info()
```

```python
#change datatypes
movie['year'].replace('TV Movie 2019', '2019', inplace = True) 
movie['year'] = movie['year'] .astype(int)
movie['date_published'].replace('TV Movie 2019', '01-01-2019', inplace = True) 
movie['date_published'] = movie['date_published'].astype('datetime64[ns]')
```

```python
# select movies published after 2000
movieAfter2000 = movie[movie['year']>=2000]
```

```python
# drop columns aren't neccessary for the rest of the project
movieAfter2000.drop(columns=['original_title','writer', 'avg_vote', 'votes', 'budget', 'usa_gross_income', 'worlwide_gross_income', 'metascore', 'reviews_from_users', 'reviews_from_critics'], inplace=True)
```

```python
# drop rows that have any nulls (NaN) in them
movieAfter2000.dropna(inplace=True)
```

```python
movieAfter2000.info()
```
